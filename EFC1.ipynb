```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w_1 = [1, 1/2, (0.5)**2, (0.5)**3, (0.5)**4]
w_2 = [1, (-0.75), 1.5, -0.2, 0.3]
```

```python
#impulso
def d(n):
  if n ==0:
    return 1
  else:
    return 0

# x[n] = s[n] − 0,5s[n − 1] + 0,07s[n − 2]
# funcao para aplicacao do impulso
def x_n(n):
  x = d(n)-0.5*d(n-1)+0.07*d(n-2)
  return x
```

```python
#definindo entrada
alphabet = np.array([1+1j,1-1j,-1+1j,-1-1j])
s = np.random.choice(alphabet,(500,))
```

```python
#aplicando o impulso para a entrada do sistema

h_n = []
for i in s:
  real = x_n(np.real(i))
  imag = x_n(np.imag(i))

  v = real +1j*imag

  h_n.append(v)

# print(h_n)
```

```python
# np.convolve
x_w1_imag = np.convolve(w_1, np.real(h_n), mode="full")
x_w1_real = np.convolve(w_1, np.imag(h_n), mode="full")
# print("y_1 = ", x_w1_imag)

x_w2_imag = np.convolve(w_2, np.real(h_n), mode="full")
x_w2_real = np.convolve(w_2, np.imag(h_n), mode="full")
# print("y_2 = ",y_2)
```

```python
#Grafico com filtro W1

plt.figure(figsize=(20,10))

# Primeiro gráfico (Parte Real)
plt.subplot(1, 2, 1)
x_axis = list(range(len(x_w1_real)))
plt.stem(x_axis, x_w1_real)
plt.xlabel('n')
plt.ylabel('x_w1_real[n]')
plt.title('Parte Real do Sinal Recebido')

# Segundo gráfico (Parte Imaginária)
plt.subplot(1, 2, 2)
x_axis = list(range(len(x_w1_imag)))
plt.stem(x_axis, x_w1_imag)
plt.xlabel('n')
plt.ylabel('x_w1_imag[n]')
plt.title('Parte Imaginária do Sinal Recebido')

# Ajusta a disposição dos gráficos para evitar sobreposição
plt.tight_layout()

# Exibe os gráficos
plt.show()
```

```python
cores1 = 'b'  # Cor para o primeiro conjunto
label1 = 's[n]'  # Rótulo para legenda

# Segundo conjunto de pontos
cores2 = 'r'  # Cor para o segundo conjunto
label2 = 'x_w1[n]'  # Rótulo para legenda

plt.figure(figsize=(20,10))

# Criação do gráfico de dispersão
plt.scatter(np.real(s), np.imag(s), color=cores1, label=label1)
plt.scatter(x_w1_real, x_w1_imag, color=cores2, label=label2)

# Configurações adicionais
plt.xlabel('Real')
plt.ylabel('Imaginário')
plt.title('Gráfico de Dispersão com Valores de Entrada e Saída')
plt.legend()

# Exibe o gráfico
plt.show()
```

```python
def ruido(lista):
  n_samples = len(lista)
  variance = 0.022
  noise = np.random.normal(0, np.sqrt(variance), n_samples)
  r = lista + noise
  return r
```

```python
#definindo saída com ruido

y1_n_real = ruido(x_w1_real)
y1_n_imag = ruido(x_w1_imag)

y2_n_real = ruido(x_w2_real)
y2_n_imag = ruido(x_w2_imag)

# print(y1_n_real)
```
